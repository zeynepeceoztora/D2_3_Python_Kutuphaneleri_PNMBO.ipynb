```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)  #-5 ile 5 arasında eşit aralıklı 100 veri noktası oluşturur.
y = np.linspace(-5, 10, 50)  #-5 ile 10 araasında eşit aralıklı 50veri noktası oluşturur.
X, Y = np.meshgrid(x, y)  # ızgara şeklini aktif ediyoruz.
Z = np.sin(np.sqrt(X**2 + Y**2))  # X ve Y üzerindeki her nokta için Z koordinatını hesaplar.
# Z değeri,, X ve Y'nin karelerinin toplamının karekökünün sinüsüdür. Bu, bir yüzey oluşturur.

ax.plot_surface(X, Y, Z, cmap='plasma')

"""
cmap komutu ile yüzey renklendirmesi sağlanır.
bone ile kemik rengi
grey ile gri rengi
ocean ile okyanus suyun mavi ton rengi
plasma ile sıcak renkler
"""

ax.set_title('3D Yüzey Grafiği Örneği')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_colormaps():
    colormaps = plt.colormaps()
    n = len(colormaps)
    fig, axs = plt.subplots(n//2, 2, figsize=(10, 2*(n//2)))
    for ax, cmap_name in zip(axs.ravel(), colormaps):
        cmap = plt.get_cmap(cmap_name)
        img = np.linspace(0, 1, 256).reshape(1, -1)
        ax.imshow(img, aspect='auto', cmap=cmap)
        ax.set_title(cmap_name, fontsize=8)
        ax.set_axis_off()
    plt.show()

plot_colormaps()
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://alicetinkaya.site/'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    links = soup.find_all('a')

    for link in links:
        href = link.get('href')
        text = link.text.strip()
        if href:
            print("Bağlantı metni: ", text, "URL:", href)
    print("Toplam bulunan bilgi sayısı: ", len(links), " adet")
else:
    print("Bağlantı başarısız: ", response.status_code)
```

```python
image = cv2.imread('Opencv/siberakademi_banner.png')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Yüklenen Görüntünün Gösterimi")
plt.show()
```

```python
# 2. aşama: görüntüyü gri tonlamaya çevirme
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.title("Gri Tonlanmış Görüntü")
plt.show()
```

```python
#3. Aşama: Görüntüyü bulanıklaştırma
blurred_image = cv2.GaussianBlur(gray_image, (5, 5),0)

#Bulanıklaştırılmış görüntüyü gösterme
plt.imshow(blurred_image, cmap='gray')
plt.title("Bulanıklaştırılmış Görüntü")
plt.show()
```

```python
#4. Aşama: Kenar algılama (Canny Edge Detection)
edges = cv2.Canny(blurred_image, threshold1=100, threshold2=200)

# Kenar algılamanın sonucunu gösterme
plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama Sonucu")
plt.show()
```

```python
#5. Aşama:
plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Orijinal Görüntü")

plt.subplot(1, 4, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Gri Tonlama")

plt.subplot(1, 4, 3)
plt.imshow(blurred_image, cmap='gray')
plt.title("Bulanık Görüntü")

plt.subplot(1, 4, 4)
plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image1 = cv2.imread('Opencv/trafik-2.png')
image2 = cv2.imread('Opencv/trafik-3.png')

hsv_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2HSV)
hsv_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV)

lower_red = np.array([0, 120, 70])
upper_red = np.array([10, 255, 255])

mask1_image1 = cv2.inRange(hsv_image1, lower_red, upper_red)
mask1_image2 = cv2.inRange(hsv_image2, lower_red, upper_red)

lower_red2 = np.array([170, 120, 70])
upper_red2 = np.array([180, 255, 255])

mask2_image1 = cv2.inRange(hsv_image1, lower_red2, upper_red2)
mask2_image2 = cv2.inRange(hsv_image2, lower_red2, upper_red2)

red_mask_image1 = mask1_image1 + mask2_image1
red_mask_image2 = mask1_image2 + mask2_image2

red_result_image1 = cv2.bitwise_and(image1, image1, mask=red_mask_image1)
red_result_image2 = cv2.bitwise_and(image2, image2, mask=red_mask_image2)

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(red_result_image1, cv2.COLOR_BGR2RGB))
plt.title('Kırmızı Ok - Görüntü 1')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(red_result_image2, cv2.COLOR_BGR2RGB))
plt.title('Kırmızı Ok - Görüntü 2')

plt.show()
```

```python
import cv2

# Resmi  yükle
img = cv2.imread('Opencv/trafik-1.png')

# Gri tonlamaya çevir
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Otsu yöntemi ile otomatik eşik bulma
ret, thresh_img = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Otsu ile bulunan eşik değeri
print(f'Otsu ile bulunan eşik değeri: {ret}')
```

```python
# Orijinal resim
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Orijinal Resim')
plt.axis('off') #eksen silme
```
